```python
# BGR to RGB
import cv2
import matplotlib.pyplot as plt

# Read image (BGR)
image = cv2.imread("sample.jpg")

# Convert to RGB
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display using Matplotlib
plt.imshow(rgb)
plt.title("RGB Image")
plt.axis("off")
plt.show()
```

```python
# Display Individual BGR Channels
import cv2

image = cv2.imread("sample.jpg")

# Split channels
blue, green, red = cv2.split(image)

cv2.imshow("Blue Channel", blue)
cv2.imshow("Green Channel", green)
cv2.imshow("Red Channel", red)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Merge BGR Channels
import cv2

image = cv2.imread("sample.jpg")

b, g, r = cv2.split(image)

merged = cv2.merge((b, g, r))

cv2.imshow("Merged Image", merged)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert to HSV
import cv2

image = cv2.imread("sample.jpg")

hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

cv2.imshow("Original", image)
cv2.imshow("HSV Image", hsv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Detect Blue Color Using HSV
import cv2
import numpy as np

image = cv2.imread("sample.jpg")

hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Blue color range
lower_blue = np.array([100, 150, 0])
upper_blue = np.array([140, 255, 255])

mask = cv2.inRange(hsv, lower_blue, upper_blue)

result = cv2.bitwise_and(image, image, mask=mask)

cv2.imshow("Original", image)
cv2.imshow("Mask", mask)
cv2.imshow("Blue Objects", result)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert to LAB
import cv2

image = cv2.imread("sample.jpg")

lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

cv2.imshow("Original", image)
cv2.imshow("LAB Image", lab)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Split LAB Channels
import cv2

image = cv2.imread("sample.jpg")

lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

L, A, B = cv2.split(lab)

cv2.imshow("Lightness", L)
cv2.imshow("A Channel", A)
cv2.imshow("B Channel", B)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert to YCrCb
import cv2

image = cv2.imread("sample.jpg")

ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

cv2.imshow("Original", image)
cv2.imshow("YCrCb", ycrcb)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Split YCrCb Channels
import cv2

image = cv2.imread("sample.jpg")

ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)

Y, Cr, Cb = cv2.split(ycrcb)

cv2.imshow("Y (Brightness)", Y)
cv2.imshow("Cr", Cr)
cv2.imshow("Cb", Cb)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert to Grayscale
import cv2

image = cv2.imread("sample.jpg")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow("Original", image)
cv2.imshow("Grayscale", gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Compare All Color Spaces
import cv2

image = cv2.imread("sample.jpg")

rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
ycrcb = cv2.cvtColor(image, cv2.COLOR_BGR2YCrCb)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imshow("Original (BGR)", image)
cv2.imshow("RGB", rgb)
cv2.imshow("HSV", hsv)
cv2.imshow("LAB", lab)
cv2.imshow("YCrCb", ycrcb)
cv2.imshow("Gray", gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Save Images in Different Color Spaces
import cv2

image = cv2.imread("sample.jpg")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

cv2.imwrite("gray.jpg", gray)
cv2.imwrite("hsv.jpg", hsv)
cv2.imwrite("lab.jpg", lab)

print("Images saved successfully.")
```

```python
# Print Pixel Values in Different Color Spaces
import cv2

image = cv2.imread("sample.jpg")

rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

x = 100
y = 150

print("BGR :", image[y, x])
print("RGB :", rgb[y, x])
print("HSV :", hsv[y, x])
print("LAB :", lab[y, x])
```
