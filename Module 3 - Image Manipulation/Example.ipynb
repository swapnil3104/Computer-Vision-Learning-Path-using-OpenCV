## Resize of image 

````python
# Resize using Width and Height
import cv2

image = cv2.imread('Sample.jpg')
if image is None:
    print("Error: Could not read the image.")
    exit()

# Resize the image to a specific width and height
resized = cv2.resize(image, (400, 300))

cv2.imshow("Original", image)
cv2.imshow("Resized", resized)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# Resize using Scale Factor
import cv2

image = cv2.imread('Sample.jpg')
if image is None:
    print("Error: Could not read the image.")
    exit()

resized = cv2.resize(image, None, fx=0.5, fy=0.5) # Resize the image to half its original size
# fx and fy are the scale factors for width and height respectively


cv2.imshow("Original", image)
cv2.imshow("Resized", resized)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# Different Interpolation Methods

import cv2

image = cv2.imread("sample.jpg")

# Resize the image using different interpolation methods
# inter_linear: Bilinear interpolation
linear = cv2.resize(image, (600, 400), interpolation=cv2.INTER_LINEAR)

# inter_area: Area-based (or pixelate) interpolation
area = cv2.resize(image, (300, 200), interpolation=cv2.INTER_AREA)

# inter_cubic: Bicubic interpolation
cubic = cv2.resize(image, (600, 400), interpolation=cv2.INTER_CUBIC)

cv2.imshow("Linear", linear)
cv2.imshow("Area", area)
cv2.imshow("Cubic", cubic)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
import cv2

image = cv2.imread("sample.jpg")

# Crop the image using array slicing
crop = image[100:350, 150:450]

cv2.imshow("Original", image)
cv2.imshow("Cropped", crop)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# Crop Center Region
import cv2

image = cv2.imread("sample.jpg")

h, w = image.shape[:2]

center = image[h//4:3*h//4, w//4:3*w//4]

cv2.imshow("Center Crop", center)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

## Rotate Image

````python
# Rotate 90 Degrees

import cv2

image = cv2.imread("sample.jpg")

# Rotate the image by 90 degrees clockwise
rotated = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

cv2.imshow("Rotated", rotated)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# Rotate 180 Degrees
import cv2

image = cv2.imread("sample.jpg")

rotated = cv2.rotate(image, cv2.ROTATE_180)

cv2.imshow("180 Rotation", rotated)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# Rotate by 45 Degrees
import cv2

image = cv2.imread("sample.jpg")

h, w = image.shape[:2]

center = (w//2, h//2)

matrix = cv2.getRotationMatrix2D(center, 45, 1)

rotated = cv2.warpAffine(image, matrix, (w, h))

cv2.imshow("45 Degree Rotation", rotated)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

## Flip Image

## Translate Image

````python
# Move Right and Down
import cv2
import numpy as np

image = cv2.imread("sample.jpg")

rows, cols = image.shape[:2]

matrix = np.float32([[1, 0, 100],
                     [0, 1, 50]])

translated = cv2.warpAffine(image, matrix, (cols, rows))

cv2.imshow("Translated", translated)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

````python
# Move Left
import cv2
import numpy as np

image = cv2.imread("sample.jpg")

rows, cols = image.shape[:2]

matrix = np.float32([[1, 0, -80],
                     [0, 1, 0]])

translated = cv2.warpAffine(image, matrix, (cols, rows))

cv2.imshow("Move Left", translated)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

## Scale Image

````python
# Enlarge Image
import cv2

image = cv2.imread("sample.jpg")

large = cv2.resize(image, None, fx=2, fy=2)

cv2.imshow("Enlarged", large)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

## Maintain Aspect Ratio

````python
import cv2

image = cv2.imread("sample.jpg")

height, width = image.shape[:2]

new_width = 500

new_height = int(height * (new_width / width))

resized = cv2.resize(image, (new_width, new_height))

cv2.imshow("Aspect Ratio Maintained", resized)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

## Compare Original and Manipulated Images

````python
import cv2

image = cv2.imread("sample.jpg")

resized = cv2.resize(image, (400, 300))

rotated = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

flipped = cv2.flip(image, 1)

cv2.imshow("Original", image)
cv2.imshow("Resized", resized)
cv2.imshow("Rotated", rotated)
cv2.imshow("Flipped", flipped)

cv2.waitKey(0)
cv2.destroyAllWindows()
````
