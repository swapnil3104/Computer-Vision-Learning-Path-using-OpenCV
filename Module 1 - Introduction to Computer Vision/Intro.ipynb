```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

print("OpenCV Version:", cv2.__version__)
```

### Reading an Image

`cv2.imread()` is used to load an image.

```python
image = cv2.imread("Dog.jpg")

if image is None:
    print("Image not found!")
else:
    print("Image Loaded Successfully")
```

### Image Information



```python
print("Image Shape :", image.shape)
print("Image Height:", image.shape[0])
print("Image Width :", image.shape[1])
print("Channels    :", image.shape[2])
```

### Display Image using Matplotlib

OpenCV stores images in **BGR** format.

Matplotlib expects **RGB** format.

Therefore, we convert BGR to RGB before displaying.

```python
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,6))
plt.imshow(rgb)
plt.axis("off")
plt.title("RGB Image")
plt.show()
```

### Display Original BGR Image

```python
plt.figure(figsize=(8,6))
plt.imshow(image)
plt.title("Without RGB Conversion")
plt.axis("off")
plt.show()
```

### Convert Image to Grayscale

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(7,5))
plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.title("Grayscale Image")
plt.show()
```

### Convert Image to HSV

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

print("HSV Shape:", hsv.shape)
```

### Split Image Channels

```python
b, g, r = cv2.split(image)

fig, ax = plt.subplots(1,3, figsize=(15,5))

ax[0].imshow(b, cmap="Blues")
ax[0].set_title("Blue Channel")
ax[0].axis("off")

ax[1].imshow(g, cmap="Greens")
ax[1].set_title("Green Channel")
ax[1].axis("off")

ax[2].imshow(r, cmap="Reds")
ax[2].set_title("Red Channel")
ax[2].axis("off")

plt.show()
```

### Save an Image

Use `cv2.imwrite()` to save processed images.

```python
cv2.imwrite("gray_image.jpg", gray)

print("Image Saved Successfully")
```
