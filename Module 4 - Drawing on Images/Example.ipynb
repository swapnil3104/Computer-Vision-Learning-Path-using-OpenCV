```python
# Create a Blank Image
import cv2
import numpy as np

# Create a white image
image = np.ones((500, 700, 3), dtype=np.uint8) * 255

cv2.imshow("Blank Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

##  Draw a Line

```python

import cv2
import numpy as np

image = np.ones((500,700,3), dtype=np.uint8) * 255

cv2.line(
    image,
    (50,100),          # Start point
    (650,100),         # End point
    (255,0,0),         # Blue
    3                  # Thickness
)

cv2.imshow("Line", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Draw Multiple Lines
import cv2
import numpy as np

image = np.ones((500,700,3), dtype=np.uint8) * 255

cv2.line(image, (50,50), (650,50), (255,0,0), 2)
cv2.line(image, (50,100), (650,200), (0,255,0), 3)
cv2.line(image, (50,250), (650,450), (0,0,255), 5)

cv2.imshow("Multiple Lines", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Draw Rectangle

```python
import cv2
import numpy as np

image = np.ones((500,700,3), dtype=np.uint8) * 255

cv2.rectangle(
    image,
    (100,100), # Top-left corner # 100,100 
    (350,250), # Bottom-right corner
    (0,255,0), # Color (Green)
    3
)

cv2.imshow("Rectangle", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Filled Rectangle
import cv2
import numpy as np

image = np.ones((500,700,3), dtype=np.uint8) * 255

cv2.rectangle(
    image,
    (150,100),
    (400,300),
    (255,0,0),
    -1
)

cv2.imshow("Filled Rectangle", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Draw Circle

```python
import cv2
import numpy as np

image = np.ones((500,500,3), dtype=np.uint8) * 255

cv2.circle(
    image,
    (250,250), # Center of the circle
    100,       # Radius
    (0,0,255), # Color (Blue)
    4
)

cv2.imshow("Circle", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Filled Circle
import cv2
import numpy as np

image = np.ones((500,500,3), dtype=np.uint8) * 255

cv2.circle(
    image,
    (250,250),
    100,
    (0,255,255),
    -1
)

cv2.imshow("Filled Circle", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Draw Multiple Circles
import cv2
import numpy as np

image = np.ones((500,700,3), dtype=np.uint8) * 255

for x in range(100, 700, 120):
    cv2.circle(image, (x,250), 50, (255,0,255), 3)

cv2.imshow("Multiple Circles", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Draw Ellipse

```python
import cv2
import numpy as np

image = np.ones((500,700,3), dtype=np.uint8) * 255

cv2.ellipse(
    image,
    (350,250),   # Center of the ellipse
    (180,90),    # Axes lengths (major and minor)
    0,           # Angle of rotation
    0,           # Start angle
    360,         # End angle
    (255,0,255), # Color (Magenta) 
    3
)

cv2.imshow("Ellipse", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Rotated Ellipse
import cv2
import numpy as np

image = np.ones((500,700,3), dtype=np.uint8) * 255

cv2.ellipse(
    image,
    (350,250),
    (180,90),
    45,
    0,
    360,
    (0,255,255),
    4
)

cv2.imshow("Rotated Ellipse", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Draw Polygon

```python
import cv2
import numpy as np

image = np.ones((500,700,3), dtype=np.uint8) * 255

points = np.array([     # Vertices of the polygon
    [100,150],     # First vertex
    [250,100],     # Second vertex
    [400,180],     # Third vertex
    [350,350],     # Fourth vertex
    [150,300]      # Fifth vertex
])

cv2.polylines(
    image,
    [points],
    True,
    (255,0,0),
    3
)

cv2.imshow("Polygon", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Filled Polygon
import cv2
import numpy as np

image = np.ones((500,700,3), dtype=np.uint8) * 255

points = np.array([
    [120,150],
    [300,80],
    [500,180],
    [420,350],
    [150,320]
])

cv2.fillPoly(
    image,
    [points],
    (0,255,0)
)

cv2.imshow("Filled Polygon", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Add Text

```python
import cv2
import numpy as np

image = np.ones((300,700,3), dtype=np.uint8) * 255

cv2.putText(
    image,
    "Hello OpenCV!",          # Text to be displayed
    (80,150),                 # Bottom-left corner of the text
    cv2.FONT_HERSHEY_SIMPLEX, # Font type
    1.5,                      # Font scale
    (0,0,255),                # Color (Red)
    3                         # Thickness
)

cv2.imshow("Text", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Write Multiple Lines of Text
import cv2
import numpy as np

image = np.ones((400,700,3), dtype=np.uint8) * 255

cv2.putText(image,"Computer Vision",(40,80),
            cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)

cv2.putText(image,"OpenCV Library",(40,170),
            cv2.FONT_HERSHEY_COMPLEX,1,(0,150,0),2)

cv2.putText(image,"Drawing Functions",(40,260),
            cv2.FONT_HERSHEY_DUPLEX,1,(0,0,255),2)

cv2.imshow("Multiple Text", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Different Fonts
import cv2
import numpy as np

image = np.ones((550,700,3), dtype=np.uint8) * 255

cv2.putText(image,"Simple",(30,60),
            cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)

cv2.putText(image,"Plain",(30,120),
            cv2.FONT_HERSHEY_PLAIN,2,(0,255,0),2)

cv2.putText(image,"Duplex",(30,180),
            cv2.FONT_HERSHEY_DUPLEX,1,(0,0,255),2)

cv2.putText(image,"Complex",(30,240),
            cv2.FONT_HERSHEY_COMPLEX,1,(255,0,255),2)

cv2.putText(image,"Triplex",(30,300),
            cv2.FONT_HERSHEY_TRIPLEX,1,(0,255,255),2)

cv2.putText(image,"Script Simplex",(30,380),
            cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,1,(50,50,50),2)

cv2.putText(image,"Script Complex",(30,460),
            cv2.FONT_HERSHEY_SCRIPT_COMPLEX,1,(150,50,200),2)

cv2.imshow("Fonts", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Draw a Simple House

```python
import cv2
import numpy as np

image = np.ones((600,700,3), dtype=np.uint8) * 255

# House
cv2.rectangle(image,(180,220),(520,500),(0,0,0),3)

# Roof
roof = np.array([[150,220],[350,70],[550,220]])

cv2.polylines(image,[roof],True,(0,0,255),3)

# Door
cv2.rectangle(image,(300,350),(400,500),(139,69,19),-1)

# Window 1
cv2.rectangle(image,(220,280),(280,340),(255,255,0),-1)

# Window 2
cv2.rectangle(image,(420,280),(480,340),(255,255,0),-1)

cv2.putText(image,
            "My House",
            (250,560),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (255,0,0),
            2)

cv2.imshow("House Drawing", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Draw an Emoji Face 😊

```python
import cv2
import numpy as np

image = np.ones((500,500,3), dtype=np.uint8) * 255

# Face
cv2.circle(image,(250,250),150,(0,255,255),-1)

# Eyes
cv2.circle(image,(200,200),20,(0,0,0),-1)
cv2.circle(image,(300,200),20,(0,0,0),-1)

# Smile
cv2.ellipse(image,
            (250,280),
            (70,40),
            0,
            0,
            180,
            (0,0,0),
            4)

cv2.imshow("Smiley", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Complete Drawing Demo

```python
import cv2
import numpy as np

image = np.ones((700,900,3), dtype=np.uint8) * 255

# Line
cv2.line(image,(40,40),(350,40),(255,0,0),3)

# Rectangle
cv2.rectangle(image,(50,80),(300,220),(0,255,0),3)

# Circle
cv2.circle(image,(500,150),70,(0,0,255),4)

# Ellipse
cv2.ellipse(image,(750,150),(80,40),30,0,360,(255,0,255),3)

# Polygon
points=np.array([[420,320],[520,260],[650,330],[600,450],[450,430]])
cv2.polylines(image,[points],True,(255,255,0),3)

# Filled Polygon
points2=np.array([[120,350],[260,320],[320,470],[150,520]])
cv2.fillPoly(image,[points2],(200,100,255))

# Text
cv2.putText(image,
            "OpenCV Drawing Functions",
            (180,650),
            cv2.FONT_HERSHEY_COMPLEX,
            1,
            (0,0,0),
            2)

cv2.imshow("Drawing Demo", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
