1) Negative of an image
The negative of an image is achieved by replacing the intensity ‘i’ in the
original image by ‘i-1’, i.e. the darkest pixels will become the brightest
and the brightest pixels will become the darkest. Image negative is
produced by subtracting each pixel from the maximum intensity value.
For example in an 8-bit grayscale image, the max intensity value is 255,
thus each pixel is subtracted from 255 to produce the output image.
The transformation function used in image negative is :
s = T(r) = (L – 1) – r
Where L - 1 is the max intensity value,
s is the output pixel value and
r is the input pixel value

```python
import cv2
import numpy as np

img=cv2.imread('landscape.jpg')
if img is not None:
    cv2.imshow('original',img)
    max1=np.max(img)
    neg_img=max1-img
    cv2.imshow('negative',neg_img)
    print(max1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Image not found or unable to load")
```

2) Log transformation
The log transformations can be defined by this formula
s = c log(r + 1).

```python
import cv2
import numpy as np

image = cv2.imread('landscape.jpg')
if image is not None:
    c = 255 / np.log(1 + np.max(image))
    log_image = c * (np.log(image + 1))
    log_image = np.array(log_image, dtype=np.uint8)
    cv2.imshow('log_img', log_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Image not found or unable to load")
```

3) Power law transformation
These transformations can be given by the expression:
s=cr^γ

```python
import cv2
import numpy as np

image = cv2.imread('landscape.jpg')
if image is not None:
    c = 255 / np.log(1 + np.max(image))
    gamma_val = 1.0
    result = c * (image ** gamma_val)
    result = np.array(result, dtype=np.uint8)
    cv2.imshow('Result', result)
    cv2.waitKey(0)
else:
    print("Image not found or unable to load")
```

4) Thresholding

```python
# Binary Thresholding
import cv2
import numpy as np

image = cv2.imread('landscape.jpg')
if image is not None:
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    ret, thresh1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    cv2.imshow("Threshold Binary", thresh1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Image not found or unable to load")
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread(r'E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\Exp 5 To Implement Point Transformations\landscape.jpg')
if image is not None:
    grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    print(grayscale.shape)

    plt.imshow(grayscale, cmap='gray')
    plt.title("Original Grayscale")
    plt.show()

    lst = []
    n = int(input("Enter number of thresholds: "))
    for i in range(n):
        ele = int(input(f"Enter threshold {i+1}: "))
        lst.append(ele)

    ht, wt = grayscale.shape
    for t in lst:
        temp = grayscale.copy()
        for i in range(ht):
            for j in range(wt):
                if temp[i][j] < t:
                    temp[i][j] = 0
                else:
                    temp[i][j] = 255
        print(f"Threshold = {t}")
        plt.imshow(temp, cmap='gray')
        plt.title(f"Thresholded at {t}")
        plt.show()
else:
    print("Image not found or unable to load")
```
