Resize Image 

```python
import cv2

img = cv2.imread("albert einstein.jpg")
Resized = cv2.resize(img , (300,300))
cv2.imshow("Image",img)
cv2.imshow("Resized Image",Resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

To convert Grayscale Image 

```python
import cv2
img = cv2.imread(r"E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 1\Demo\albert einstein.jpg")

# Convert to Grayscale
Gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Image",img)
cv2.imshow("Grayscaled Image ",Gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Crop the image 

```python
import cv2
img = cv2.imread(r"E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 1\Demo\albert einstein.jpg")

# Crop a region of image 
Cropped_img = img[100:300, 200:400]

cv2.imshow("Image",img)
cv2.imshow("Cropped Image",Cropped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
