```python
import cv2
```

```python
# Program to display an image using OpenCV
# path = r'E:\COLLEGE NOTES\SEM 7th Notes\Practical\DIP\EXP 1/Lenna.png'
img=cv2.imread('albert einstein.jpg', cv2.IMREAD_UNCHANGED)
cv2.imshow('Albert Einstein', img)
cv2.waitKey(0)  # Wait indefinitely until any key is pressed
```

```python
# Program to rotate an image
src = cv2.imread('albert einstein.jpg', cv2.IMREAD_UNCHANGED)
window_name = 'image'
image = cv2.rotate(src, cv2.ROTATE_90_CLOCKWISE)
cv2.imshow(window_name, image)
cv2.waitKey(0)
```

```python
# Program to Flip an image 

src = cv2.imread('albert einstein.jpg', cv2.IMREAD_UNCHANGED)
flip_vertical = cv2.flip(src, 0)
flip_horizontal = cv2.flip(src, 1)
flip_both = cv2.flip(src, -1)

cv2.imshow('Original Image', src)
cv2.imshow('Vertical Flip', flip_vertical)
cv2.imshow('Horizontal Flip', flip_horizontal)
cv2.imshow('Both Flip', flip_both)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Program to Transpose an image 

src = cv2.imread('albert einstein.jpg', cv2.IMREAD_UNCHANGED)
transposed = cv2.transpose(src)

cv2.imshow('Original Image', src)
cv2.imshow('Transposed Image', transposed)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Program to convert original image to grayscale

src = cv2.imread('albert einstein.jpg', cv2.IMREAD_UNCHANGED)
gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

cv2.imshow('Original Image', src)
cv2.imshow('Grayscale Image', gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
