To create a program to rotate an image.

```python
import cv2

path = r"E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 1\albert einstein.jpg"
src = cv2.imread(path)

if src is None:
    print("Error: Could not open or find the image.")
else:
    # Rotating the image 90 degrees clockwise
    image = cv2.rotate(src, cv2.ROTATE_90_CLOCKWISE)
    
 
    window_name = 'Rotated Image'
    cv2.imshow(window_name, image)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

To create a program to flip an image in Separate Frame 

```python
import cv2

path = r"E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 1\albert einstein.jpg"
originalImage = cv2.imread(path)

if originalImage is None:
    print("Error: Could not read the image.")
else:
    
    flipVertical = cv2.flip(originalImage, 0)   
    flipHorizontal = cv2.flip(originalImage, 1) 
    flipBoth = cv2.flip(originalImage, -1)      


    cv2.imshow('Original image', originalImage)
    cv2.imshow('Flipped vertical image', flipVertical)
    cv2.imshow('Flipped horizontal image', flipHorizontal)
    cv2.imshow('Flipped both image', flipBoth)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

 All Flip Image in One Frame 

```python
import cv2

path = r"E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 1\albert einstein.jpg"
image = cv2.imread(path)

if image is None:
    print("Error: Image not found")
else:
    
    flipVertical = cv2.flip(image, 0)   
    flipHorizontal = cv2.flip(image, 1) 
    flipBoth = cv2.flip(image, -1)      
 
    width, height = 300, 300
    img1 = cv2.resize(image, (width, height))
    img2 = cv2.resize(flipVertical, (width, height))
    img3 = cv2.resize(flipHorizontal, (width, height))
    img4 = cv2.resize(flipBoth, (width, height))

    top_row = cv2.hconcat([img1, img2])
    bottom_row = cv2.hconcat([img3, img4])
    combined = cv2.vconcat([top_row, bottom_row])

    
    cv2.imshow('All Images in One Frame', combined)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

To create a program to convert original image to grayscale.

```python
import cv2

path = r"E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 1\albert einstein.jpg"
image = cv2.imread(path)

if image is None:
    print("Error: Could not read the image.")
else:
    # Convert image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    cv2.imshow('Original Image', image)
    cv2.imshow('Grayscale Image', gray_image)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
