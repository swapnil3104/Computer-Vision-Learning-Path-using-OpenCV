```python
# Importing Necessary Libraries
from skimage import data, filters
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

# Step 1: Load a sample image (coffee image from skimage)
coffee = data.coffee()

# Step 2: Convert the image to grayscale
gray_coffee = rgb2gray(coffee)

# Step 3: Set up a large figure for displaying multiple images
plt.figure(figsize=(15, 15))

# Step 4: Loop through different threshold values (0.0 to 0.9)
for i in range(10):
    # Apply manual thresholding — convert grayscale to binary
    binarized_gray = (gray_coffee > i * 0.1) * 1

    # Plot each thresholded image in a 5x2 grid
    plt.subplot(5, 2, i + 1)
    plt.title("Threshold: >" + str(round(i * 0.1, 1)))
    plt.imshow(binarized_gray, cmap='gray')

plt.tight_layout()
plt.show()
```

```python
# Importing necessary libraries
from skimage import data, filters
from skimage.color import rgb2gray
import matplotlib.pyplot as plt

# Setting plot size
plt.figure(figsize=(15, 15))

# Step 1: Load the sample image
coffee = data.coffee()

# Step 2: Convert the image to grayscale
gray_coffee = rgb2gray(coffee)

# Step 3: Otsu’s Global Thresholding
threshold = filters.threshold_otsu(gray_coffee)
binarized_coffee = (gray_coffee > threshold) * 1
plt.subplot(2, 2, 1)
plt.title("Otsu Threshold: > " + str(round(threshold, 3)))
plt.imshow(binarized_coffee, cmap="gray")

# Step 4: Niblack Local Thresholding
threshold = filters.threshold_niblack(gray_coffee)
binarized_coffee = (gray_coffee > threshold) * 1
plt.subplot(2, 2, 2)
plt.title("Niblack Thresholding")
plt.imshow(binarized_coffee, cmap="gray")

# Step 5: Sauvola Local Threshold (Not Binarized)
threshold = filters.threshold_sauvola(gray_coffee)
plt.subplot(2, 2, 3)
plt.title("Sauvola Threshold Map")
plt.imshow(threshold, cmap="gray")

# Step 6: Sauvola Local Threshold (Binarized)
binarized_coffee = (gray_coffee > threshold) * 1
plt.subplot(2, 2, 4)
plt.title("Sauvola Thresholding - Binary")
plt.imshow(binarized_coffee, cmap="gray")

plt.tight_layout()
plt.show()
```
