```python
import cv2

# Read image
src = cv2.imread('albert einstein.jpg')

# Check if image is loaded
if src is not None:
    print(src[100, 100][0])  # Blue channel value
    print(src[100, 100][1])  # Green channel value
    print(src[100, 100][2])  # Red channel value
else:
    print("Error: Could not read the image.")
```

```python
import cv2


path = r'E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 2\albert einstein.jpg'
src = cv2.imread(path)

#  Red channel
src[:, :, 0] = 0   # Set Blue channel to 0
src[:, :, 1] = 0   # Set Green channel to 0


cv2.imshow("Red Channel Image", src)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

For Green Channel

```python
import cv2

path = r'E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 2\albert einstein.jpg'
src = cv2.imread(path)

if src is None:
    print("Error: Could not read the image. Check file path or name.")
else:
    # Green channel
    green_img = src.copy()
    green_img[:, :, 0] = 0   # Blue
    green_img[:, :, 2] = 0   # Red

    cv2.imshow("Green Channel", green_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

For Blue Channel

```python
import cv2

path = r'E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 2\albert einstein.jpg'
src = cv2.imread(path)

# Blue channel
blue_img = src.copy()
blue_img[:, :, 1] = 0   # Green
blue_img[:, :, 2] = 0   # Red

cv2.imshow("Blue Channel", blue_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Negative Image 

```python
import cv2


path = r'E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 2\albert einstein.jpg'
src = cv2.imread(path)

if src is None:
    print("Error: Could not read the image. Check the path.")
else:
    # Create negative image
    negative_img = 255 - src

    # Display images
    cv2.imshow("Original Image", src)
    cv2.imshow("Negative Image", negative_img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

Gray Scale Image

```python
import cv2


path = r'E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 2\albert einstein.jpg'
src = cv2.imread(path)

if src is None:
    print("Error: Could not read the image. Check the path.")
else:
    # Convert to grayscale
    gray_img = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)


    cv2.imshow("Original Image", src)
    cv2.imshow("Grayscale Image", gray_img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

HSV Image 

```python
import cv2


path = r'E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 2\albert einstein.jpg'
src = cv2.imread(path)

if src is None:
    print("Error: Could not read the image. Check the path.")
else:
    # Convert BGR image to HSV
    hsv_img = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)

    # Display images
    cv2.imshow("Original Image", src)
    cv2.imshow("HSV Image", hsv_img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

YcbCr Image 

```python
import cv2

path = r'E:\COLLEGE NOTES\SEM 7th Notes\Practical\Digital Image Processing\EXP 2\albert einstein.jpg'
src = cv2.imread(path)

if src is None:
    print("Error: Could not read the image. Check the path.")
else:
    # Convert BGR image to YCbCr
    ycbcr_img = cv2.cvtColor(src, cv2.COLOR_BGR2YCrCb)

    cv2.imshow("Original Image", src)
    cv2.imshow("YCbCr Image", ycbcr_img)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
