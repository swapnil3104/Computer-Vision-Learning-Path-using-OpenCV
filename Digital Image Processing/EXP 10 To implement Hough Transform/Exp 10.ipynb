```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Read the image in grayscale
img = cv2.imread('lotus.jpg', 0)
cv2.imshow("Input image", img)
cv2.waitKey(0)

# Step 2: Apply Gaussian Blur to reduce noise
img = cv2.GaussianBlur(img, (3, 3), 0)

# Step 3: Perform Canny edge detection
canny1 = cv2.Canny(img, 50, 150)
cv2.imshow('Canny', canny1)
cv2.waitKey(0)

# Step 4: Detect lines using Probabilistic Hough Transform
lines_list = []
lines = cv2.HoughLinesP(canny1, 1, np.pi / 180, threshold=50,
                        minLineLength=5, maxLineGap=20)

# Step 5: Draw the detected lines
for points in lines:
    x1, y1, x2, y2 = points[0]
    cv2.line(img, (x1, y1), (x2, y2), (0, 0, 165), 2)
    lines_list.append([(x1, y1), (x2, y2)])

cv2.imshow('Detected Lines', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
