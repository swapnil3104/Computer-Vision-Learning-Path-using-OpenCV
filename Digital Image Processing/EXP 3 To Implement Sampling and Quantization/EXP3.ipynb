Resize of image

```python
import cv2

img = cv2.imread("albert einstein.jpg")
if img is None:
    print("Error: Could not read the image.")
else:
    Resized = cv2.resize(img, (300, 300))
    cv2.imshow("Original Image", img)
    cv2.imshow("Resized Image", Resized)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

Percentage Resize

```python
import cv2

img = cv2.imread("albert einstein.jpg")
if img is None:
    print("Error: Could not read the image.")
else:
    height = img.shape[0]
    width = img.shape[1]
    percent = 50
    w = int(width * percent / 100)
    h = int(height * percent / 100)
    resize = cv2.resize(img, (w, h))
    print(resize.shape)
    cv2.imshow("Resize", resize)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

Image Pyramid using OpenCV

Image Pyramids are one of the most beautiful concepts of image processing.
Normally, we work with images with default resolution but many times we
need to change the resolution (lower it) or resize the original image in that
case image pyramids comes handy.
The pyrUp() function increases the size to double of its original size and
pyrDown() function decreases the size to half. If we keep the original image
as a base image and go on applying pyrDown function on it and keep the
images in a vertical stack, it will look like a pyramid. The same is true for
upscaling the original image by pyrUp function.

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("albert einstein.jpg")
if img is None:
    print("Error: Could not read the image.")
else:
    layer = img.copy()
    for i in range(4):
        plt.subplot(2, 2, i + 1)
        layer = cv2.pyrDown(layer)
        plt.imshow(cv2.cvtColor(layer, cv2.COLOR_BGR2RGB))
        plt.title(f"Level {i+1}")
        plt.axis("off")

        cv2.imshow(str(i), layer)
        cv2.waitKey(0)

    cv2.destroyAllWindows()
    plt.show()
```

Image quantization

PIL is the Python Imaging Library which provides the python interpreter with
image editing capabilities. The Image module provides a class with the same
name which is used to represent a PIL image. The module also provides a
number of factory functions, including functions to load images from files,
and to create new images.
Image.quantize() Convert the image to ‘P’ mode with the specified number
of colors.

Syntax: Image.quantize(colors=256, method=None, kmeans=0,palette=None)

```python
from PIL import Image
import PIL

img = Image.open("albert einstein.jpg")
img = img.quantize(170)
img.show()
```
