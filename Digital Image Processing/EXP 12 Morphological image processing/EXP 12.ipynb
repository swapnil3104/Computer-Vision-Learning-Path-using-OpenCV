```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image in grayscale
img = cv2.imread(r"images1.jpg", 0)

# Step 2: Apply Otsu's binarization
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Step 3: Create a kernel for morphological operations
kernel = np.ones((5, 5), np.uint8)

# Step 4: Invert the binary image (so black <-> white)
invert = cv2.bitwise_not(binr)

# Step 5: Perform erosion on the inverted image
erosion = cv2.erode(invert, kernel, iterations=1)

# Step 6: Display the result
plt.imshow(erosion, cmap='gray')
plt.axis('off')
plt.show()
```

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image in grayscale
img = cv2.imread(r"images2.jpg", 0)

# Step 2: Apply Otsu's binarization
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Step 3: Define a 3x3 kernel (structuring element)
kernel = np.ones((3, 3), np.uint8)

# Step 4: Perform morphological opening (Erosion → Dilation)
opening = cv2.morphologyEx(binr, cv2.MORPH_OPEN, kernel, iterations=1)

# Step 5: Display the result
plt.imshow(opening, cmap='gray')
plt.axis('off')
plt.show()
```

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image in grayscale
img = cv2.imread(r"images1.jpg", 0)

# Step 2: Apply Otsu's thresholding
# NOTE: cv2.threshold() returns two values — threshold and image.
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Step 3: Define the kernel (structuring element)
kernel = np.ones((3, 3), np.uint8)

# Step 4: Invert the binary image
invert = cv2.bitwise_not(binr)

# Step 5: Apply Morphological Gradient
# MORPH_GRADIENT = Dilation - Erosion
morph_gradient = cv2.morphologyEx(invert, cv2.MORPH_GRADIENT, kernel)

# Step 6: Display the result
plt.imshow(morph_gradient, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# Import the necessary packages
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image in grayscale
img = cv2.imread(r"images2.jpg", 0)

# Step 2: Apply Otsu's thresholding
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Step 3: Define the kernel (13x13 for larger neighborhood)
kernel = np.ones((13, 13), np.uint8)

# Step 4: Apply Top-hat transformation
# MORPH_TOPHAT = Original Image - Opening
morph_tophat = cv2.morphologyEx(binr, cv2.MORPH_TOPHAT, kernel)

# Step 5: Display the result
plt.imshow(morph_tophat, cmap='gray')
plt.axis('off')
plt.show()
```

```python

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image in grayscale
img = cv2.imread(r"images2.jpg", 0)

# Step 2: Apply Otsu's thresholding
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Step 3: Define the kernel (structuring element)
kernel = np.ones((5, 5), np.uint8)

# Step 4: Invert the binary image (optional depending on feature direction)
invert = cv2.bitwise_not(binr)

# Step 5: Apply Morphological Black-hat transformation
# MORPH_BLACKHAT = Closing - Original
morph_blackhat = cv2.morphologyEx(invert, cv2.MORPH_BLACKHAT, kernel)

# Step 6: Display the result
plt.imshow(morph_blackhat, cmap='gray')
plt.axis('off')
plt.show()
```
