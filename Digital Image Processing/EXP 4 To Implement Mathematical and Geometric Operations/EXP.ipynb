```python
import cv2

image1 = cv2.imread('Space.jpg')
image2 = cv2.imread('astronaut.jpg')
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))
cv2.imshow("First Image - Space", image1)
cv2.imshow("Second Image - Astronaut (Resized)", image2)
addition = cv2.add(image1, image2)
cv2.imshow("Added Image", addition)
blend = cv2.addWeighted(image1, 0.7, image2, 0.3, 0) 
cv2.imshow("Blended Image", blend)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

3) Multiplication of Image:
We can multiply the pixel values in two images and merge them with
the help of cv2.multiply(). The images should be of equal size and
depth. 

```python
import cv2
image1 = cv2.imread('Space.jpg')
image2 = cv2.imread('astronaut.jpg')
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))
mul = cv2.multiply(image1, image2)
cv2.imshow("First Image", image1)
cv2.imshow("Second Image", image2)
cv2.imshow("Multiplied Image", mul)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

1) AND Operation:
To perform bitwise AND logical operations, Python OpenCV
provides the cv2.bitwise_and() method. It combines the
corresponding pixels of two image buffers using a bitwise AND
operation. It is commonly used for detecting differences in input
images. It returns an image highlighting the target regions with a
binary mask.

```python
import cv2

image1 = cv2.imread('Airplane.jpg')
image2 = cv2.imread('sky.jpeg')
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))
bitwise_and = cv2.bitwise_and(image1, image2)
cv2.imshow("First Image", image1)
cv2.imshow("Second Image", image2)
cv2.imshow("Bitwise AND", bitwise_and)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

2) OR Operation
Python OpenCV provides the cv2.bitwise_or() method to perform
bitwise OR logical operation. It combines the corresponding pixels
of two image buffers by a bitwise OR operation. The bitwise OR
operator is useful for processing binary-valued images (0 or 1) to
detect objects that have moved between frames. Here is an example

```python
import cv2

image1 = cv2.imread('Airplane.jpg')
image2 = cv2.imread('sky.jpeg')
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))
bitwise_or = cv2.bitwise_or(image1, image2)
cv2.imshow("Airplane", image1)
cv2.imshow("Sky", image2)
cv2.imshow("Bitwise OR", bitwise_or)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

4) NOT operation
Bitwise NOT or complement is a unary operation that performs
logical negation on each bit. It forms the one's complement of the
given binary value. It means the bits that are 1 become 0, and those
that are 0 become 1. Python OpenCV provides the cv2.bitwise_not
() method to perform a bitwise NOT operation on each pixel of the
input image. It inverts the image representation.

```python
import cv2

image = cv2.imread('Airplane.jpg')
bitwise_not = cv2.bitwise_not(image)
cv2.imshow("Original Image", image)
cv2.imshow("Bitwise NOT", bitwise_not)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
